```python
import os 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
os.chdir("C:\DataScience_DSC_640\Weeks3-4\ex2-2")
```

```python
world_pop_df = pd.read_excel("world-population.xlsm")
```

```python
world_pop_df.head(3)
```

```python
# Line Chart 

plt.figure(figsize=(10,6))

plt.plot(world_pop_df['Year'],world_pop_df['Population'])
plt.title("World Population Growth Over The Years")
plt.xlabel("Years")
plt.ylabel("Population in Billions")

plt.show()
```

```python
# Step Chart 

plt.figure(figsize=(10,6))

plt.step(world_pop_df['Year'],world_pop_df['Population'])
plt.title("World Population Growth Over The Years")
plt.xlabel("Years")
plt.ylabel("Population in Billions")

plt.show()
```
